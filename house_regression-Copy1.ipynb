```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
num_house = house.loc[:,df.dtypes==np.int64]

calc_vif(num_house)
```

```python

pd.options.display.float_format = '{:,.2f}'.format
calc_vif(num_house).sort_values(by='VIF', ascending=True)
```

```python
house_dummy = pd.get_dummies(house,drop_first=True)
```

```python
house_dummy['GrLivArea'] = house_dummy['GrLivArea'].apply(lambda x: np.log(x+1))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(house_dummy, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline import make_pipeline 
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
#training model using pipline that way we get more operations at once.
Pipeline=make_pipeline(StandardScaler(), SGDClassifier())
est=Pipeline.fit(X_train,y_train)
prediction=est.predict(X_test)
```

```python
import numpy as np
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, prediction)) #Mean absolute error
print('MSE:', metrics.mean_squared_error(y_test, prediction))#Mean squared error (punishes outliers)
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```

```python
from sklearn.linear_model import Lasso
Pipeline2=make_pipeline(StandardScaler(), Lasso(alpha=0.01, max_iter=1000000))
est2=Pipeline2.fit(X_train,y_train)
prediction=est2.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction)) #Mean absolute error
print('MSE:', metrics.mean_squared_error(y_test, prediction))#Mean squared error (punishes outliers)
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```

```python
from sklearn.linear_model import Ridge
Pipeline_ridge=make_pipeline(StandardScaler(), Ridge())
est5=Pipeline_ridge.fit(X_train,y_train)
prediction=est5.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction)) #Mean absolute error
print('MSE:', metrics.mean_squared_error(y_test, prediction))#Mean squared error (punishes outliers)
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```

```python
from sklearn.linear_model import ElasticNet
Pipeline_ela=make_pipeline(StandardScaler(), ElasticNet())
est4=Pipeline_ela.fit(X_train,y_train)
prediction=est4.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction)) #Mean absolute error
print('MSE:', metrics.mean_squared_error(y_test, prediction))#Mean squared error (punishes outliers)
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid_ela ={'alpha': [0.1, 0.5, 1, 2, 5], 
                 'l1_ratio': [0, 0.25, 0.5, 0.75, 1]}

optimal_param_ela = GridSearchCV(estimator = ElasticNet(max_iter=5000), param_grid = param_grid_ela, scoring='neg_root_mean_squared_error')
optimal_param_ela.fit(X_train,y_train)
optimal_param_ela.best_estimator_
```
